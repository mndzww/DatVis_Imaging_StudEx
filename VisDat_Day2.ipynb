```python
import matplotlib.pyplot as plt
#example data
x=[0,1,2,3,4,5]
y=[0,1,4,9,16,25]

#create the plot
plt.plot(x,y,label="x squared", color="blue",marker='o')

#add title an labels
plt.title("Simple line graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

#Show grid
plt.grid(True)

#Display the legend
plt.legend()

#Show the plot
plt.show
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    "X1": [71, 39, 34, 76, 84, 64, 100, 73, 71, 39, 83, 93, 26, 15],
    "X2": [52, 61, 64, 70, 68, 47, 37, 54, 41, 55, 70, 36, 26, 42],
    "X3": [58, 40, 70, 43, 55, 52, 59, 56, 41, 50, 65, 74, 74, 98],
    "X4": [47, 82, 42, 70, 62, 72, 66, 80, 52, 65, 75, 97, 97, 38],
}

df = pd.DataFrame(data)
df.to_csv("data_sampling.csv", index=False)

styled_df = df.style.set_table_styles(
    [{"selector": "th", "props": [("border", "1px solid black"), ("background-color", "F0FFF0")]},
     {"selector": "td", "props": [("border", "1px solid black")]}]
)
display(styled_df)
```

```python
mean_values = df.mean().to_frame(name="Mean")
median_values = df.median().to_frame(name="Median")
mode_values = df.mode().iloc[0].to_frame(name="Mode")

display(mean_values)
display(median_values)
display(mode_values)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(7, 5))

# Subplot 1: X1 vs X4
axes[0, 0].scatter(df["X1"], df["X4"], c='blue', label="X1 vs X4")
axes[0, 0].set_xlabel("X1")
axes[0, 0].set_ylabel("X4")
axes[0, 0].set_title("Scatter Plot X1 vs X4")
axes[0, 0].legend()
axes[0, 0].grid(True)

# Subplot 2: X2 vs X3
axes[0, 1].scatter(df["X2"], df["X3"], c='red', label="X2 vs X3")
axes[0, 1].set_xlabel("X2")
axes[0, 1].set_ylabel("X3")
axes[0, 1].set_title("Scatter Plot X2 vs X3")
axes[0, 1].legend()
axes[0, 1].grid(True)

# Subplot 3: X1 vs X3
axes[1, 0].scatter(df["X1"], df["X3"], c='green', label="X1 vs X3")
axes[1, 0].set_xlabel("X1")
axes[1, 0].set_ylabel("X3")
axes[1, 0].set_title("Scatter Plot X1 vs X3")
axes[1, 0].legend()
axes[1, 0].grid(True)

# Subplot 4: X2 vs X4
axes[1, 1].scatter(df["X2"], df["X4"], c='purple', label="X2 vs X4")
axes[1, 1].set_xlabel("X2")
axes[1, 1].set_ylabel("X4")
axes[1, 1].set_title("Scatter Plot X2 vs X4")
axes[1, 1].legend()
axes[1, 1].grid(True)

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8, 6))
max_value = df.values.max()
bins = np.arange(0, max_value + 15, 15)  

plt.figure(figsize=(8, 5))
plt.hist(df.values.flatten(), bins=bins, edgecolor='black', alpha=0.7)

plt.xticks(bins) 
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
plt.title("Histogram Distribusi Data Sampling")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df["X1"], df["X4"], marker='o', linestyle='-', color='green', label="X1 vs X4")
plt.xlabel("X1")
plt.ylabel("X4")
plt.title("Kurva XY (X1 vs X4)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap Korelasi Data Sampling")
plt.show()
```
